```python
# ตั้งชื่อ field ใหม่ทั้งหมด
new_column_names = ['ที่', 'กรมธรรม์', 'ชื่อ-สกุล', 'ที่อยู่1', 'วันที่จ่ายเงิน', 'ปี','จำนวนเงิน' ,'รับเอง' ,'รับมอบฯ' ,'null','รหัสสาขา']
df.columns = new_column_names
# แทนที่ค่า null ด้วยตัวก่อนหน้า

df['กรมธรรม์'] = df['กรมธรรม์'].astype(str)
length = df['กรมธรรม์'].str.len()
df['กรมธรรม์'] = df['กรมธรรม์'].where(length >= 6, df['กรมธรรม์'].shift())
# เปลี่ยน type ข้อมูล
dtypes = {
        "ที่": "str",
        "กรมธรรม์": "str",
        "รหัสสาขา": "str",
        "ชื่อ-สกุล": "object",
        "ที่อยู่1": "object", 
        # "ที่อยู่2": "object",
        "วันที่จ่ายเงิน": "object",
        "ปี": "object",
        "จำนวนเงิน": "str",
        "รับเอง": "object",
        "รับมอบฯ": "object"
    }
df = df.astype(dtypes)
# ตัดแถวว่างออก
df = df[df['ที่อยู่1'] . notnull()]
df.head()

```

```python
# เอาที่เลขกรมธรรม์เป็น null ออก เอาไว้ไปเติมข้อมูล
null_df = df[df['วันที่จ่ายเงิน'] . isnull()]
null_df.head()
null_field = ['กรมธรรม์' , 'ที่อยู่1']
null_df = null_df[null_field]
# เปลี่ยนชื่อ field
null_df.columns = ['กรมธรรม์' , 'ที่อยู่2']
# null_df.head()
```

```python
# ดึงเลขกรมธรรม์ที่ไม่ null
notnull_df = df[df['วันที่จ่ายเงิน'] .notnull()]
# notnull_df.head()
```

```python
# ่join data
data_df = pd.merge(notnull_df, null_df, on='กรมธรรม์', how='left')

# เรียงลำดับคอลัมน์ใหม่ 
column_order = ['ที่', 'กรมธรรม์','รหัสสาขา' , 'ชื่อ-สกุล', 'ที่อยู่1', 'ที่อยู่2', 'วันที่จ่ายเงิน', 'ปี','จำนวนเงิน' ,'รับเอง' ,'รับมอบฯ' ,'null']
data_df = data_df[column_order]

# ตัด culumn null ออก
data_df = data_df[['ที่', 'กรมธรรม์','รหัสสาขา' , 'ชื่อ-สกุล', 'ที่อยู่1', 'ที่อยู่2', 'วันที่จ่ายเงิน', 'ปี','จำนวนเงิน' ,'รับเอง' ,'รับมอบฯ']]

data_df.head()
```

```python
# ตัดแถวที่ไม่ต้องการ
data_df_finish = data_df[data_df['ปี'] .notnull()]
# notnull_df = df[df['กรมธรรม์'] .notnull()]
# data_df_finish.head()
```

```python
data_df_finish['จำนวนเงิน'] = data_df_finish['จำนวนเงิน'].str.replace(',', '').astype(int)
data_df_finish = data_df_finish.sort_values(by=['รหัสสาขา', 'จำนวนเงิน'], ascending=[True, False])
```

#### ส่วน ACL

```python
Rule_IND_df = data_df_finish[data_df_finish['รับมอบฯ'] == '/']
# เรียงลำดับ รหัสสาขา กับ จำนวนเงิน 
Rule_IND_df = Rule_IND_df.sort_values(by=['รหัสสาขา', 'จำนวนเงิน'], ascending=[True, False])
Rule_IND_df.head()
```

##### Export

```python
# วันที่ปัจจุบัน
now = datetime.now().strftime("%Y%m%d")

# แยก DataFrame ตามรหัสสาขาและบันทึกไฟล์ Excel
grouped = Rule_IND_df.groupby('รหัสสาขา')

for รหัสสาขา, group in grouped:
    # จำกัดจำนวน record ไม่เกิน 5
    if len(group) > 5:
        group = group.iloc[:5]
    record_count = group.shape[0]  # นับจำนวน record หลังการแบ่งไฟล์
    output_file_name = f"Rule2_OPsurrenderIND_{รหัสสาขา}_รับมอบ_{now}_{record_count}.xlsx"
    file_path = os.path.join(path, output_file_name)
    group.to_excel(file_path, index=False)
```
